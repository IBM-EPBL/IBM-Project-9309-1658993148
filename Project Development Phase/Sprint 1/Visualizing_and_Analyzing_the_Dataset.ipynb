```python
#bivariate analysis
df.plot()
```

```python
sns.catplot(data=df, x="having_IPhaving_IP_Address", y="Result")
```

```python
sns.scatterplot(data=df, x="URLURL_Length", y="Result")
```

```python
sns.barplot(df.having_At_Symbol.value_counts().index, df.Result.value_counts())
```

```python
sns.lineplot(df['having_At_Symbol'],df['Result'])
```

```python
def countplot_of_2(x,hue,title=None,figsize=(6,5)):
    plt.figure(figsize=figsize)
    sns.countplot(data=df[[x,hue]],x=x,hue=hue)
    plt.title(title)
    plt.show()
```

```python
countplot_of_2('having_At_Symbol','Result','having_At_Symbol vs Result')
```
